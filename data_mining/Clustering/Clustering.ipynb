# Clustering

El clustering (o agrupamiento) es una técnica de data mining o machine learning que sirve para agrupar datos similares entre sí dentro de un conjunto más grande homogeneo.

<p align="center">
  <img src="img/clustering.png" width="600">
</p>

Su objetivo es que se de una division de datos de la siguiente forma:

- Los objetos dentro de un mismo grupo (cluster) sean similares (intra-cluster).
- Los objetos de diferentes grupos sean diferentes entre sí (inter-cluster).

Es una técnica no supervisada, lo que significa que no hay etiquetas o clases predefinidas. El algoritmo busca estructuras ocultas o patrones en los datos por sí mismo.

El clustering se utilizaa en variedad de aplicaciones como:

- La segmentacion de clientes y adaptar estrategias para cada grupo.
- Reduccion de dimensionalidad, cambio de tener datos con x parametros, los evaluamos a partir de sus k clases.
- La deteccion de anomalias, evitando fraudes.
- Aprendizaje semisupervisado, primeramente agrupar por clustering y darles una etiqueta

### Principales algoritmos de clustering

| Algoritmo                         | Características principales                                                                                                                                   |
| --------------------------------- | ------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **K-Means**                       | Divide los datos en *k* grupos. Usa la media (centroide) de los puntos del cluster. Es rápido pero sensible a la posición inicial y a la escala de los datos. |
| **Hierarchical Clustering**       | Crea una jerarquía de grupos (como un árbol o dendrograma). No requiere elegir *k* al inicio.                                                                 |
| **DBSCAN**                        | Detecta clusters de forma arbitraria y puntos "ruido" (outliers). Muy útil para datos espaciales o con densidad variable.                                     |
| **Mean-Shift**                    | Agrupa los puntos moviéndose hacia las zonas de mayor densidad. No requiere definir *k*, pero es más costoso computacionalmente.                              |
| **Gaussian Mixture Models (GMM)** | Supone que los datos provienen de una mezcla de distribuciones normales. Más flexible que K-Means.                                                            |
